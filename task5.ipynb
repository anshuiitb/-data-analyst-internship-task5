# 1. IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```


# Display full outputs

```python
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
```

# 2. LOAD DATASET

```python
# Using seaborn Titanic dataset
df = sns.load_dataset('titanic')
```

# 3. BASIC EXPLORATION

```python
print("First 5 rows:")
display(df.head())

print("\nDataset Info:")
print(df.info())

print("\nSummary Statistics:")
display(df.describe(include='all'))

print("\nMissing Values:")
print(df.isnull().sum())
```

# 4. UNIVARIATE ANALYSIS

```python
plt.figure(figsize=(8,5))
sns.histplot(df['age'].dropna(), kde=True, bins=30, color='skyblue')
plt.title("Age Distribution")
plt.show()

# Observation: Most passengers are between 20-40 years.

# Count of survivors
plt.figure(figsize=(6,4))
sns.countplot(x='survived', data=df, palette='Set2')
plt.title("Survival Count")
plt.show()

# Observation: More passengers did not survive.
```

# 5. BIVARIATE ANALYSIS

```python
# Survival rate by gender
plt.figure(figsize=(6,4))
sns.countplot(x='sex', hue='survived', data=df)
plt.title("Survival by Gender")
plt.show()

# Observation: Females had higher survival rate.

# Boxplot: Age vs Class
plt.figure(figsize=(6,4))
sns.boxplot(x='class', y='age', data=df, palette='pastel')
plt.title("Age Distribution per Class")
plt.show()

# Observation: First class tends to have older passengers.
```

# 6. MULTIVARIATE ANALYSIS

```python
# Correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# Observation: Fare and Pclass have strong correlation.

# Pairplot
sns.pairplot(df.dropna(subset=['age','fare']), vars=['age','fare'], hue='survived', palette='husl')
plt.show()
```


## Observation: Survived and non-survived groups overlap in fare-age distribution.##

  SUMMARY OF FINDINGS
- Majority did not survive (~62%).
- Females had significantly higher survival rates than males.
- Higher class passengers (First class) had higher survival rate.
- Fare is positively correlated with survival and class.
- Most passengers aged 2040 years.
